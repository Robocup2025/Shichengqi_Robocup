### Canny Detection手动

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('raw.jpg', 0)

blurred = cv2.GaussianBlur(img, (5, 5), 1.4)

grad_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
grad_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

magnitude = np.sqrt(grad_x**2 + grad_y**2)
magnitude_uint8 = cv2.convertScaleAbs(magnitude)

nms_emulation = cv2.Canny(blurred, 50, 100) 

final_edges = cv2.Canny(blurred, 100, 200)

def show_canny_steps(step1, step2, step3):
    plt.figure(figsize=(18, 6))
    
    plt.subplot(1, 3, 1)
    plt.imshow(step1, cmap='gray')
    plt.title('1. Gradient Magnitude ')
    plt.axis('off')
    
    plt.subplot(1, 3, 2)
    plt.imshow(step2, cmap='gray')
    plt.title('2. After NMS ')
    plt.axis('off')
    
    plt.subplot(1, 3, 3)
    plt.imshow(step3, cmap='gray')
    plt.title('3. Edge Linking ')
    plt.axis('off')
    
    plt.tight_layout()
    plt.show()

show_canny_steps(magnitude_uint8, nms_emulation, final_edges)
```
