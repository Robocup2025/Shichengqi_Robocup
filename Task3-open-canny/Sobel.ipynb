### 手动实现Sobel算子

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def sobel_demo(img):
    # 计算x, y方向梯度
    grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
    grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
    # 合并梯度
    grad = cv2.addWeighted(cv2.convertScaleAbs(grad_x), 0.5, cv2.convertScaleAbs(grad_y), 0.5, 0)
    return grad

# 读取并处理
img = cv2.imread('raw.jpg', 0) # 请确保路径下有图片
res1 = sobel_demo(img)
plt.imshow(res1, cmap='gray'), plt.title('Sobel Gradient'), plt.show()
```
